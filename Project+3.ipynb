```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('DataCleaned.csv')
df
```

```python
df.dtypes
```

## Test #1

```python
# Step #1 Select the fields
cols = ['population', 'avg_household_size', 'avg_car_amount_apt', 'avg_car_amount_house', 'families_with_children_pct', 'married_pct', 'single_females_pct', 'single_males_pct']
# I chose these fields because I wanted to see how population of these neighborhoods affected the other fields I selected.I included these fields because I felt that population could actually have an impact on these fields, instead of it just being a coincidence.
```

```python
# Step #2 Normalize the data
X = StandardScaler().fit_transform(df[cols])
X
```

```python
# Step #3 Determine how many clusters you need
max_clusters = 2
squared_distance = np.zeros(max_clusters)
for k in range(2, max_clusters):
    kmeans = KMeans(n_clusters = k, random_state = 0).fit(X)
    squared_distance[k] = kmeans.inertia_
   # print(k,sep=' ', end=' ', flush=True)
    
plt.figure()
plt.plot(squared_distance)
plt.xlim((2, max_clusters))
plt.xlabel('Clusters Needed')
plt.ylabel('Distance From Center')
plt.show()

# The highest amount of clusters I can add is two, so I will only use two clusters
```

```python
# Step #4 Run Clustering
km = KMeans(n_clusters = 2, random_state = 0).fit(X)
km.fit(df[cols])

df['cluster'] = km.labels_

df.groupby('cluster')[cols].mean().sort_values(by='population')
```
